## Base64 Endcoding

  very old method:
     https://facebook/com/search/<userID>/photos-by
    
  next old method:
     https://facebook/com/search/photos/?q=photos&epa=FILTERS&filters={"rp_author:{\"name\":\"author\","args\":\"[USERID]\"}"}
     
  new method base64:
     https://facebook/com/search/photos/?q=photos&epa=FILTERS&filters=[EQUIVALENT_BASE64_REPRESENTATION]
     
  Here are sites to generate and de-generate:
     https://codebeautify.org/base64-encode/
     https://codebeautify.org/base64-decode/
     
     from conversion get hash with trailing ==    (final two-equals "==" are optional, and not necessary
